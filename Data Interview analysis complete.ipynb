```python
# Visualize the correlation of data
import seaborn as sns
sns.heatmap(df.corr())
```

```python
# plot line graph of Date against Bet Amount
sns.lineplot(x=df['Date'],y=df['Bet Amount'], data=df)
plt.title('bet amount over time')
plt.xlabel('Date')
plt.ylabel('bet Amount');
```

Graph above shows bet amont has been on a steady uptrend from March 2018 towards the end of January 2019 where it starts to decline.

```python
#plot line graph of Date against current period profit
sns.lineplot(x=df['Date'],y=df['Current Period Profit'], data=df)
plt.title('current period profit over time')
plt.xlabel('Date')
plt.ylabel('current period profit');
```

Graph above show the current period profit has been on an uptrend until mid March 2019 from where a steady decline is observed.

```python
#plot line graph of Date against gift amount
sns.lineplot(x=df['Date'],y=df['Gift Amount'], data=df)
plt.title('Gift amount over time')
plt.xlabel('Date')
plt.ylabel('Gift Amount');
```

From the above visualization we can see that mid may the gift amount has been on a series of downtrend till the end of the period.

```python
# Plot to compare bet amount vs winning amount
sns.lineplot(x=df['Date'],y=df['Bet Amount'], data=df,color='navy', label='Bet Amount')
sns.lineplot(x=df['Date'],y=df['Winning Amount'], data=df,color='teal', label='Winning Amount')
plt.title('Betting Amount Compared to Winning Amount')
plt.xlabel('Date')
plt.ylabel('Amount');
```

The graph above shows that betting amount is positively correlated to the winning amount. 

```python
sns.lineplot(x=df['Date'],y=df['Current Period Profit'], data=df,color='navy', label='Current Period Profit')
sns.lineplot(x=df['Date'],y=df['Revenue Calculation'], data=df,color='teal', label='Revenue Calculation')
plt.title('Current Period Profit Compared to Revenue Calculation')
plt.xlabel('Date')
plt.ylabel('Amount');
```

from the graph above we can deduce that current period profit is positively correlated with revenue calculation.

```python
sns.lineplot(x=df['Date'],y=df['Bet Amount'], data=df,color='navy', label='Bet Amount')
sns.lineplot(x=df['Date'],y=df['Gift Amount'], data=df,color='teal', label='Gift Amount')
plt.title('Betting Amount Compared to gift Amount')
plt.xlabel('Date')
plt.ylabel('Amount');
```

From the graph above we can conclude that gift amount does not influence the bet amount.
